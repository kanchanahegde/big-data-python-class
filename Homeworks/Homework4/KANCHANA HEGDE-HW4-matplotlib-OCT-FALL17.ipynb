#### Homework 4.2
 

Make a combined plot of a  red spiral with a scatter plot.

The red spiral has r from 0 to 3 with steps of 0.1 and
a theta of 2 * pi * r.

The scatter uses the same theta as the spiral but the radius r is offset by a random value ranging from -.5 to .5.
The size of the plotted locations is 100 * r^2 and the color uses theta.

Use an alpha value of .4 for the colors in the scatter plot.
Limit the size of the plot (rmax) to 2.5
 
 Should look something like:
<img src="./images/lineplot_with_scatterplot.png" width="300" height="200">


#### Steps to create above graphs

1. Steps to create red spiral
    1. create x and y for red spiral
       y=r, 300 points inthe range of 0 to 3
       x=theta, using equation, 2 * pi * r
    2. create a subplot and use projection as 'polar'
    3. plot the theta, r in the above subplot to get red spiral
    
    
2. Steps to create spiral bubbles
    1. since spiral bubbles have same circumference as red spiral, let us use same above theta
    2. for r, we will add some offset by creating 300 random numbers betwwen 0 to .5 using linspace function or           uniform function from random package
    3. using scatter plot, plot the graph in the same subplot to get the result.

```python
import random
r = np.arange(0, 3, 0.01)
theta = 2 * np.pi * r
ax = plt.subplot(111,projection='polar')
ax.plot(theta,r,'r')

# adding offset using linspace function
t=np.linspace(0, .5, len(theta))
random.shuffle(t)
r=r+t

ax.scatter(theta,r,c=theta,s=100*r**2,alpha=.4,cmap='jet',edgecolors='k')
ax.set_rticks([0.5, 1, 1.5, 2,2.5])  # less radial ticks
ax.set_rmax(2.5)
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```

```python
import random
r = np.arange(0, 3, 0.01)
theta = 2 * np.pi * r
ax = plt.subplot(111,projection='polar')
ax.plot(theta,r,'r')

# adding offset using uniform function
t=np.random.uniform(low=0, high=.5, size=(300,))
r=r+t

ax.scatter(theta,r,c=theta,s=100*r**2,alpha=.4,cmap='jet',edgecolors='k')
ax.set_rticks([0.5, 1, 1.5, 2,2.5]) 
ax.set_rmax(2.5)
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```

Create a 2d histogram of 100000 points

x is a generated from a normal random distribution 

y is generated from a geometric random value witha p=.01 
Use the LogNorm color settings for hist2d.

Should look something like:

#### Steps to create above graphs

1. import LogNorm which normalizes the data via log 10.
2. create a figure of size (5,4)
3. create x and y for hist2d function
   1. x= create 100000 random normally distributed points.
   2. y= using Geometric distribution create output same as x and with probability p = 0.01
4. plot the data using hist2d function , with data tranformed by logNorm form to base 10 to get the above result.

```python
from matplotlib.colors import LogNorm
from pylab import *
fig=plt.figure(figsize=(5,4))
#normal distribution center at x=0 and y=5
x = np.random.randn(100000)
y = np.random.geometric(p=0.01, size=100000)

plt.hist2d(x, y, bins=40, norm=LogNorm(),cmap='jet')
plt.colorbar(ticks=[0, 10,100,1000])
plt.title('2d Color Histogram')
show()
```

#### Steps to create above graphs

1. Create a subplot with project as '3d'
2. Using the same x and y created in the previous example,Compute the bi-dimensional histogram of two data.
   1. hist = bi-dimensional histogram of samples x and y.
   2. X = Values in x are histogrammed along the first dimension.
   3. Y = Values in y are histogrammed along the second dimension.
3. Create a rectangular grid out of an array of X values and an array of Y values using numpy meshgrid function
4. Now using the plot_surface function, plot the histogram on 3d surface.



```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.randn(100000)
y = np.random.geometric(p=0.01, size=100000)

hist,x,y = np.histogram2d(x, y, bins=40)
# Express the mesh in the cartesian system.
# Remove last row,as the size returned by histogram2d is for X and Y is 1 more than initial the x and y.
X,Y = np.meshgrid(x[:-1] , y[:-1] )

# Plot the surface.
ax.plot_surface(X, Y, hist ,rstride=1,cstride=1,cmap=plt.cm.jet,edgecolors='k')
ax.set_title('3D plot of 2D histogram')

plt.show()
```
